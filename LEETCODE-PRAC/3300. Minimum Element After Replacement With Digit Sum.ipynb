```python
#brutrforce approach
class Question:
    def minElement(self, nums: List[int]) -> int:
        nums_split = [str(_) for _ in nums]
        i = 0
        while(i!=len(nums)):
            for x in nums_split:
                dig_sum = 0
                for _ in range(len(x)):
                    dig_sum = dig_sum + int(x[_])
                nums[i] = int(dig_sum)
                i = i + 1
        return(min(nums))    
```

```python
#optimal approach
class Solution:
    def minElement(self, nums: List[int]) -> int:
        mini = float("inf")
        for n in nums:
            curr = 0
            while(n>0):
                d = n%10
                curr += d
                n = n//10
            mini = min(mini,curr)
        return mini
```

```python
ans = Solution()
ans.minElement([96,89])
```
